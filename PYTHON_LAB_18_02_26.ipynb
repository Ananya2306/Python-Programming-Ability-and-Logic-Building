- wapip - for cal the electricity bill  for 1st 100 units the unit cost should be - 1 100 unit = rupees 3 per unit, next 100 units - 5 per unit, above 200 units 7 per units

```python
#- wapip - for cal the electricity bill  for 1st 100 units the unit cost should be - 1 100 unit = rupees 3 per unit, next 100 units - 5 per unit, above 200 units 7 per units
units = int(input())
if units <= 100:
  bill = units * 3
elif units>=100 and units <= 200:
  bill = 100 * 3 + (units - 100) * 5
else:
  bill = (units * 3) + (units-100) * 5 + (units - 200) *7
print(bill)
```
