```python
# to print factorial of given number using function
num = int(input("Enter a number: "))
def factorial(n):
  if n == 0 or n==1:
    return 1
  else:
    return n * factorial(n-1)
result = factorial(num)
print(f"The factorial of {num} is {result}")
```

```python
# to delete an element in the list using func
def delete_element_from_list(my_list, element_to_delete):
  try:
    my_list.remove(element_to_delete)
    print(f"Element '{element_to_delete}' removed successfully.")
  except ValueError:
    print(f"Element '{element_to_delete}' not found in the list.")
  return my_list

my_list = [10, 20, 30, 40, 50, 20]
my_list = delete_element_from_list(my_list, 20)
print(f"List after deleting 20: {my_list}")
```

```python
#write a function to check the factorial of given number using iterators
n = int(input("Enter the number:- "))
def fact(n):
  if n < 0:
    return "Factorial is not defined for negative numbers"
  elif n == 0 or n == 1:
    return 1
  else:
    res = 1
    for i in range(1, n + 1):
      res *= i
    return res

result = fact(n)
print(f"The factorial of {n} is {result}")
```
